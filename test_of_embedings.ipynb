```python
from ultralytics import YOLO
import numpy as np
```

```python
img1 = 'gos1.png'
img2 = 'gos2.png'
img3 = './chrome.png'
```

```python
model = YOLO('manuspect_models/baseline_yolov8s.pt')
model.info()
```

```python
res = model.predict(img1)
v1 = np.array(res[0].embedding)
print(v1)
```

```python
res = model.predict(img2)
v2 = np.array(res[0].embedding)
print(v2)
```

```python
res = model.predict(img3)
v3 = np.array(res[0].embedding)
print(v3)
```

```python
distance_gos_gos = np.linalg.norm(v1 - v2)
distance_gos_chrome = np.linalg.norm(v3 - v2)

print("The distance between the gos uslugi imgages is:", distance_gos_gos)

print("The distance between the gos uslugi ang chrome imgages is:", distance_gos_chrome)
```

```python
res = model.predict(img3)
v4 = np.array(res[0].embedding)
res = model.predict(img3)
v5 = np.array(res[0].embedding)

distance_for_same_img = np.linalg.norm(v5 - v4)
print("The distance between the same images is:", distance_for_same_img)
```
